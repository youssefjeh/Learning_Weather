```python
import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import re
```

```python
dates_r= pd.date_range(start="01/01/2010",end="01/01/2022",freq="M")
dates = [str(i)[:4] + str(i)[5:7] for i in dates_r]
lst = []
index = []
for j in range(len(dates)):
    url="https://www.estesparkweather.net/archive_reports.php?date="+ dates[j] 
    page=requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    table = soup.find_all('table')
    data_rows= [row.text.splitlines() for row in table]
    data_rows= data_rows[:-9] 
    for k in range(len(data_rows)):
        data_rows[k] = data_rows[k][2:len(data_rows[k]):3]
    for l in range(len(data_rows)):
        c= [('.'.join(re.findall("\d+",str(data_rows[l][k].split()[:5])))) for k in range(len(data_rows[l]))]
        lst.append(c)
        index.append(dates[j] + c[0])
d1_index = [index[i] for i in range(len(index)) if len(index[i]) > 6]
data = [lst[i][1:] for i in range(len(lst)) if len(lst[i][1:]) == 19]
```

```python
print(d1_index)
```

```python
final_index = [datetime.strptime(str(d1_index[i]), "%Y%m%d").strftime("%Y-%m-%d") for i in range(len(d1_index))]
final_index
```

```python
weather_data=pd.DataFrame(data)
```

```python
weather_data.insert(0,"date",final_index,allow_duplicates=False)
```
